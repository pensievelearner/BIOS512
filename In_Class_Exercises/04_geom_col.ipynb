```r
options(repr.plot.width=6, repr.plot.height=2.5, repr.plot.res = 300)

# write your chart code below 👇
p = ggplot(cases, aes(x = date, y = positiveIncrease))

p  = p + geom_col(width = 1)
# This makes it so that the gaps in the chart are filled: (width = 1)

p = p + scale_y_continuous(labels = comma)

# He found the above online to create commas for the y axis

p
```
