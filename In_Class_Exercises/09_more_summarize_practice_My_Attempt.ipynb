Use `filter` and `summarize` to produce two single row tables. One that has the median `temp` for January, and one that has the median `temp` for June.

```r
# E.g.

median.january  = weather %>% 
    group_by(month) %>%
    filter(month == 1) %>% 
    summarize(median.temp = median(temp))

median.june = weather %>%
        group_by(month) %>%
            filter(month == 6) %>%
                summarize(median.temp = median(temp))

median.january
median.june
```
