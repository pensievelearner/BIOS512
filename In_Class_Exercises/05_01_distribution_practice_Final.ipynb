The code below will pull out the vaccination rate for North Carolina.

````r
vax_data %>%
    filter(`State/Territory/Federal Entity` == "North Carolina") %>%
    pull(`Percent of 18+ Pop with at least One Dose by State of Residence`) %>%
    paste("of 18 or older North Carolinians have at least 1 dose")
````

Can you use `geom_vline` to add a vertical line representing the position of NC's vaccination rate to your histogram? What arguments does `geom_vline` take?

````r
p = ggplot(vax_data, aes(x = `Percent of 18+ Pop with at least One Dose by State of Residence`))

p = p + geom_histogram()

p = p + geom_rug()

p = p +  geom_vline(xintercept = 69.4)

p
````

````r
vax_data_long = vax_data %>% 
    rename(perc_18plus_one_dose = `Percent of 18+ Pop with 1+ Doses by State of Residence`,
           perc_18plus_two_dose = `Percent of 18+ Pop with 2 Doses by State of Residence`) %>%
    #colnames - to check it worked - after we saw it worked we continued to the below line
#We used the below to select the columns we wanted
    select(perc_18plus_one_dose, perc_18plus_two_dose, `State/Territory/Federal Entity`) %>%
#This is the code to make data in long form
    pivot_longer(cols= c("perc_18plus_one_dose","perc_18plus_two_dose"), 
                 values_to = 'percent', 
                 names_to = 'dose_number') %>%
    mutate(nc = ifelse(`State/Territory/Federal Entity` == 'North Carolina', 'yes', 'no'))
        # When you transpose the default table has columns with the names "value" and "name" we chose to rename
        # this with the following code appended to the end in the lines above--> values_to = 'percent', names_to = 'dose_number'

````

````r
p3 = ggplot(vax_data_long, aes(x = percent, y = dose_number, color = nc))

p3 = p3 + geom_point(position = position_jitter(w = 0, h = 0.2)) 
    # We add the jitter because if you look at it without it there are a bunch of points squashed together and jitter helps 
    # spread it out so you can more clearly see the data

p3 = p3 + scale_color_manual(values = c("gray60", "firebrick"))
                             
p3
````

````r
# A student asked if we could have so that NC always stayed on top of the other points
# This lead to a discussion about why it's difficult. Jitter randomly scatters the points while following the parameters you specify

p4 = ggplot(vax_data_long, aes(x = percent, y = dose_number, color = nc))

p4 = p4 + geom_point(position = position_jitter(w = 0, h = 0.2), color = 'grey70',
                    data = vax_data_long %>% filter(nc == 'no') )

# Now we try to superimpose the NC point on top of all the others

p4 = p4 + geom_point(position = position_jitter(w = 0, h = 0.2), color = 'red', size = 2.5,
                    data = vax_data_long %>% filter(nc == 'yes') )

#p4 = p4 + scale_color_manual(values = c("gray60", "firebrick"))
                             
p4
````

````r
# We spent time in lecture talking about ggrepel to add labels to points and now we are revisiting the above plot
# Need to load library before doing anything else
library('ggrepel') 
````

````r
p5 = ggplot(vax_data_long, aes(x = percent, y = dose_number, color = nc, label = `State/Territory/Federal Entity`))
    # We added the label to use in geom_text_repel

p5 = p5 + geom_point(position = position_jitter(w = 0, h = 0.2), color = 'grey70',
                    data = vax_data_long %>% filter(nc == 'no') )

p5 = p5 + geom_point(position = position_jitter(w = 0, h = 0.2), color = 'red', size = 2.5,
                    data = vax_data_long %>% filter(nc == 'yes') )


p5 = p5 + geom_text_repel(data = vax_data_long %>% filter(nc == 'yes'), segment.color = "grey50",
                            nudge_x = -50)

p5

# I am not sure why my North Carolina is Red - his was black
````
