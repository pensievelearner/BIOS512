Can you make this plot? ðŸ‘‡ What does this plot demonstrate?

```r
p3 = ggplot(penguins, aes(x = body_mass_g, y = flipper_length_mm, color = species))

p3 = p3 + geom_point(na.rm = TRUE) 

# na.rm tells r to ignore the warnings
# p2 = p2 + scale_color_brewer(type = "qual", palette = "Set2")

p3
```

```r
# Goal to create a facet grid of these plots
options (repr.plot.width = 5, repr.plot.height = 5)

p5 = ggplot(penguins, aes(x = body_mass_g, y = flipper_length_mm))

p5= p5 + geom_point(na.rm = TRUE) 

p5 = p5 + facet_grid(island~species)

p5
```

```r
# Goal to create a stacked bar chart
# Also the p4 grid got deleted

options (repr.plot.width = 5, repr.plot.height = 4)

p6 = ggplot(penguins, aes(y = island, fill = species))

p6= p6 + geom_bar(na.rm = TRUE) 

p6= p6+ scale_fill_brewer(palette = "Set2")

p6
```
