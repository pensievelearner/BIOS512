Use `geom_bar` to make a histogram of penguin counts by `year`.

```r
p = ggplot(penguins, aes(y = year))

p = p + geom_bar(fill = 'lightblue', color = 'black')

p = p + scale_x_continuous(name = 'Number of penguins') + ylab('Year data recorded')

p
```

Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
p2 = ggplot(penguins, aes(x = body_mass_g))

p2 = p2 + geom_bar(color= 'black', fill = 'lightblue')

p2 = p2 + facet_wrap( ~year,nrow = 3, ncol = 1)

p2 = p2 + scale_x_continuous(name = 'Body mass in grams')

p2 = p2 + scale_y_continuous(name = 'Number of penguins', limit=c(0,7.5))

p2= p2 + theme_light()

#p = p + scale_y_continuous(labels = comma)

p2
```

🎉**BONUS**: can you adjust the `x-axis` labels to have "comma" formatting? (**Hint:** use `comma` from the `scales` package.)

```r
p3 = ggplot(penguins, aes(x = body_mass_g))

p3 = p3 + geom_bar(color= 'black', fill = 'lightblue')

p3 = p3 + facet_wrap( ~year,nrow = 3, ncol = 1)

p3 = p3 + scale_x_continuous(name = 'Body mass in grams', label = scales::comma)

p3 = p3 + scale_y_continuous(name = 'Number of penguins', limits = c(0, 7.5))

p3 = p3 + theme_light()

#p = p + scale_y_continuous(labels = comma)

p3
```

Is there a difference between the body mass of male and female penguis? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to see compare within each species-group.**

```r
penguins %>% head
```

```r
p4 = ggplot(penguins, aes(x = body_mass_g, y = sex))

p4 = p4 + geom_point(position = position_jitter(w = 0.2, h = 0.3), alpha = 0.3) 

p4 = p4 + facet_wrap( ~species,nrow = 3, ncol = 1)

p4 = p4 + scale_x_continuous(name = 'Body mass in grams', label = scales::comma)

p4 = p4 + scale_y_discrete(name = 'Sex of penguin')

p4 = p4 + theme_light()

p4
```
