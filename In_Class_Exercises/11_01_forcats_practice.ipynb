```r
median_mpg = mpg %>% 
      group_by(class) %>%
        summarize(cty = median(cty)) %>%
            arrange(cty) %>%
                mutate(class = fct_inorder(class))

# He didn't use median_cty he actually kept cty because he intentionally wants it to overlap when you combine the two plots
```

```r
p = ggplot(median_mpg, aes( x = cty, y = class))

p = p + geom_crossbar(xmin=0, xmax = 0, color = 'blue')

p
```
