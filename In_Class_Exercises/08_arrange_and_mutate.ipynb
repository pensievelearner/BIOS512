```r
library('tidyverse')
```

```r
df.marathon = read.csv('../data//marathon.csv')
df.marathon %>% head
```

Use `arrange` to find the best race time.

```r
df.marathon %>%
        arrange(race.time) %>%
            head(1) #This only outputs the first row
```

What are the best race times by country?

```r
df.marathon %>%
        group_by(country) %>% 
            summarize(Best_Time = min(race.time)) %>%
                arrange(Best_Time)
```

```r
df.marathon %>%
        group_by(country) %>% 
            arrange(race.time) %>%
                mutate(rk = row_number()) %>%
                    filter(rk ==1)
                           
                           
                        
```

In what year was the difference between the men's and women's winning time smallest? Use this table:

```r
df = df.marathon %>%
    select(year, MF, race.time) %>% # remember that if you want to exclude anything you would put a "-" in front of it
    tidyr::spread(MF, race.time) %>% # tidyr :: spread means that we are taking the spread function from tidyr even if we haven't formally loaded it at the top
    drop_na() %>% # We drop the missing values because we are only taking the difference of true values
    rename(men = `Men's`, women = `Women's`) # we renamed because having to type `Men's` everytime is irritating

df %>% head
```

```r
new = df %>%
    mutate(M_W_Diff = abs(women - men)) %>%
        group_by(year) %>%
            arrange(M_W_Diff)
```

Plot the difference between men's and women's race times by year.

```r
options(repr.plot.res = 250, repr.plot.width = 5, repr.plot.height = 4)

p = ggplot(new, aes(x = M_W_Diff, y = year))

p = p + geom_point()

p
```
